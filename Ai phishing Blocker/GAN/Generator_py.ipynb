**Code to Genetrete Synthatic urls.:**

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LeakyReLU

# Define the Generator model
def build_generator(domain_length):
    model = Sequential()
    model.add(Dense(128, input_dim=100))
    model.add(LeakyReLU(alpha=0.01))
    model.add(Dense(256))
    model.add(LeakyReLU(alpha=0.01))
    model.add(Dense(domain_length, activation='sigmoid'))
    return model

# Get user input URL
user_input_url = input("Enter a URL: ")

# Extract domain name from user input URL
domain_name = user_input_url.split('//')[1].split('.')[0]

# Convert domain name to a format compatTible with model input
domain_length = len(domain_name)
domain_dataset = np.array([list(domain_name.encode())])

# Build the generator model
generator = build_generator(domain_length)

# Generate domain names and print in the same URL structure
def generate_and_print_domains(n_samples, generator, domain_length, original_url):
    noise = np.random.normal(0, 1, (n_samples, 100))
    generated_domains = generator.predict(noise)
    print(generated_domains[0])
    generated_domains_list = [
        "".join([chr(int(val * 26 + 97)) for val in domain]) # Convert values to letters
        for domain in generated_domains
    ]

    # Print generated domains with the same URL structure
    for generated_domain in generated_domains_list:
        generated_url = original_url.replace(domain_name, generated_domain)
        print(generated_url)
generate_and_print_domains(n_samples=100, generator=generator, domain_length=domain_length, original_url=user_input_url)
```
